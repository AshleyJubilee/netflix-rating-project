```python
import pandas as pd
```

```python
titleData = pd.read_csv("../resources/titles.csv")
titleData.head()
```

```python
titleData.describe()
```

```python
showData = pd.DataFrame(titleData.loc[titleData["type"].values == ["SHOW"], ["id", "title", "release_year", "age_certification", "runtime", "genres", "production_countries", "seasons", "imdb_score"]])
showData.reset_index(inplace=True, drop=True)
showData.head()
```

```python
movieData = pd.DataFrame(titleData.loc[titleData["type"].values == ["MOVIE"], ["id", "title", "release_year", "age_certification", "runtime", "genres", "production_countries", "imdb_score"]])
movieData.reset_index(inplace=True, drop=True)
movieData.head()
```

```python
showData.to_csv('../resources/showData.csv')
```

```python
movieData.to_csv('../resources/movieData.csv')
```
