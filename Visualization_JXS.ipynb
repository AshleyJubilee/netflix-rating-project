```python
# Explode the 'genres' column to create separate rows for each genre
df_exploded = df_shows.explode('genres')

# Create a new DataFrame to count the occurrences of each genre in each year
genre_counts = df_exploded.groupby(['release_year', 'genres']).size().unstack(fill_value=0)

# Calculate the total popularity of each genre across all years
genre_counts_total = genre_counts.sum()

# Select the top 10 genres based on total popularity
top_10_genres = genre_counts_total.nlargest(10).index

# Plotting the graph
plt.figure(figsize=(12, 6))

# Iterate through each genre and plot its popularity over the years if it's in the top 10
for genre in top_10_genres:
    plt.plot(genre_counts.index, genre_counts[genre], label=genre)

plt.title('Popularity of Top 10 TV Show Genres Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of TV Shows')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Group data by release_year and age_certification and count occurrences
certification_counts = df_shows.groupby(['release_year', 'age_certification']).size().reset_index(name='count')

# Plot the data
plt.figure(figsize=(12, 8))
sns.lineplot(data=certification_counts, x='release_year', y='count', hue='age_certification', marker='o')
plt.title('Evolution of Age Certification of Shows Over the Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Shows')
plt.xticks(rotation=45)
plt.legend(title='Age Certification')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Group data by release_year and age_certification and count occurrences
certification_counts = df_shows.groupby(['release_year', 'age_certification']).size().unstack(fill_value=0)

# Plot the data
certification_counts.plot(kind='area', stacked=True, figsize=(12, 8))
plt.title('Evolution of Age Certification of Shows Over the Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Shows')
plt.xticks(rotation=45)
plt.legend(title='Age Certification')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Explode the genres column to have one row per genre
df_genres = df_movies.explode('genres')

# Count occurrences of each genre
genre_counts = df_genres['genres'].value_counts()

# Get top 10 genres
top_10_genres = genre_counts.head(10).index

# Filter the data for top 10 genres
top_10_data = df_genres[df_genres['genres'].isin(top_10_genres)]

# Count occurrences of each genre per year for top 10 genres
top_10_genre_counts = top_10_data.groupby(['release_year', 'genres']).size().unstack(fill_value=0)

# Plotting
plt.figure(figsize=(12, 8))

# Plot each genre as a separate line
for genre in top_10_genre_counts.columns:
    plt.plot(top_10_genre_counts.index, top_10_genre_counts[genre], label=genre)

plt.title('Popularity of Top 10 Movie Genres Over the Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Movies')
plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Group by release year and age certification and count the occurrences
certification_counts = df_movies.groupby(['release_year', 'age_certification']).size().unstack(fill_value=0)

# Plot
certification_counts.plot(kind='area', stacked=True, figsize=(12, 8), colormap='tab20')
plt.title('Evolution of Age Certification of Movies Over the Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Movies')
plt.legend(title='Age Certification')
plt.grid(True)
plt.tight_layout()
plt.show()
```
